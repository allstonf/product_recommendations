## Web Scraping using BeautifulSoup

```python
url = "https://www.amazon.com/s?k=snow+sled&ref=nb_sb_noss_1" #"https://www.amazon.com/s?k=snow+sled&page=2&qid=1595968264&ref=sr_pg_2" #"http://www.amazon.com/s/ref=sr_in_-2_p_4_18?me=A3MZ96G5C78IVQ&fst=as%3Aoff&rh=p_4%3AFunKo&ie=UTF8&qid=1477811368"
print("running cell")

# add header
headers = {
    'User-Agent': 
#     'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)' 
    'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'
}

print("making GET request")
r = requests.get(url, headers=headers)
print(r)
# dr = webdriver.Chrome()
# dr.get(url)
# print(dr)

print("creating BeautifulSoup object\n")
soup = BeautifulSoup(r.content, "lxml")
print(soup.prettify())
# soup = BeautifulSoup(dr.page_source,"lxml")
# print(soup.prettify())

print("opening file")
file = open(r"parseddata.txt", "w")

print("finding all links")
links = soup.find_all('a', {'class': 
                            "a-link-normal a-text-normal" #'a-link-normal s-access-detail-page a-text-normal'
                           })

for link in links:
    print(link.get('href'))
    file.write(link.get('href')+ '\n')
file.close()
```
